# PDF to Image to Text

This notebook is for taking all the PDF files and exporting the text into .txt files. OCR is used as this gives better results than using direct extraction with tools such as pdfminer, and because in can be applied to scanned documents. The PDFs are converted into images, which are scanned with pytesseract to extract the text in them. Finally, the text are exported into.txt files

```python
from pdf2image import convert_from_path
from pytesseract import image_to_string
import pytesseract
from PIL import Image
```

```python
# If you don't have tesseract executable in your PATH, include the following:
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

def convert_pdf_to_img(pdf_file):
    """
    @desc: this function converts a PDF into Image
    
    @params:
        - pdf_file: the file to be converted
    
    @returns:
        - an interable containing image format of all the pages of the PDF
    """
    return convert_from_path(pdf_file, poppler_path=r"C:\Users\alejo\Documents\Maestria_en_Ciencia_de_Datos\Trabajo_de_Grado\poppler-24.02.0\Library\bin")


def convert_image_to_text(file, lang):
    """
    @desc: this function extracts text from image
    
    @params:
        - file: the image file to extract the content
        - lang: language of the document. Make sure that the necessary language data is installed in Tesseract-OCR
    
    @returns:
        - the textual content of single image
    """
    
    text = image_to_string(file, lang)
    return text


def get_text_from_any_pdf(pdf_file, lang):
    """
    @desc: this function is our final system combining the previous functions
    
    @params:
        - file: the original PDF File
        - lang: language of the document. Make sure that the necessary language data is installed in Tesseract-OCR
    
    @returns:
        - the textual content of ALL the pages
    """
    images = convert_pdf_to_img(pdf_file)
    final_text = ""
    for _pg, img in enumerate(images):
        
        final_text += convert_image_to_text(img, lang)
        #print("Page n°{}".format(pg))
        #print(convert_image_to_text(img))
    
    return final_text
```
